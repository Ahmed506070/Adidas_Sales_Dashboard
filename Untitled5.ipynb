```python
df = pd.read_excel("Adidas US Sales Datasets.xlsx", header=4)
df.head()
```

```python
df.describe()
```

```python
print(f"zeroo: {(df['Total Sales']==0).sum()}")
```

```python
df = df[df['Total Sales'] > 0]
```

```python
print(f" الداتا نظيفة : {df.shape[0]:,} صف")
```

```python
df['Year'] = df['Invoice Date'].dt.year
```

```python
df['Month'] = df['Invoice Date'].dt.month
```

```python
print(df[['Invoice Date','Year','Month']].head())
```

```python
total_sales = df['Total Sales'].sum()
```

```python
total_units = df['Units Sold'].sum()
```

```python
avg_margin = df['Operating Margin'].mean()
```

```python
total_profit = df['Operating Profit'].sum()
```

```python
print(f"💰 إجمالي المبيعات    : ${total_sales:,.0f}")
print(f"📈 إجمالي الأرباح     : ${total_profit:,.0f}")
print(f"📉 متوسط هامش الربح  : %{avg_margin*100:.1f}")
print(f"📦 إجمالي الوحدات     : {total_units:,}")
```

```python
region_data = (df.groupby('Region')[['Total Sales', 'Operating Profit']]
               .sum()
               .sort_values('Total Sales', ascending=False)
               .reset_index())

print(region_data)
```

```python
fig, ax = plt.subplots()

x = range(len(region_data))
w = 0.4

ax.bar([i - w/2 for i in x], region_data['Total Sales'] / 1e6,
       width=w, label='Total Sales', color='#4C72B0')

ax.bar([i + w/2 for i in x], region_data['Operating Profit'] / 1e6,
       width=w, label='Operating Profit', color='#55A868')

ax.set_xticks(list(x))
ax.set_xticklabels(region_data['Region'])
ax.set_ylabel('Sales (Million $)')
ax.set_title('Total Sales & Operating Profit by Region')
ax.legend()

plt.tight_layout()
plt.savefig('chart_regions.png', dpi=150)
plt.show()
```

```python
product_data = (df.groupby('Product')[['Total Sales', 'Operating Profit']]
                .sum()
                .sort_values('Total Sales', ascending=False)
                .reset_index())

print(product_data)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

sns.barplot(data=product_data,
            y='Product',
            x='Total Sales',
            palette='Set2',
            ax=ax)

ax.set_title('Total Sales by Product Category')
ax.set_xlabel('Total Sales ($)')
ax.set_ylabel('Product')

plt.tight_layout()
plt.savefig('chart_products.png', dpi=150)
plt.show()
```

```python
method_data = (df.groupby('Sales Method')[['Total Sales', 'Operating Profit']]
               .sum()
               .reset_index())

print(method_data)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Pie Chart - توزيع المبيعات
axes[0].pie(
    method_data['Total Sales'],
    labels=method_data['Sales Method'],
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('Set2', len(method_data))
)
axes[0].set_title('Sales Distribution by Method')

# Bar Chart - الأرباح
sns.barplot(data=method_data,
            x='Sales Method',
            y='Operating Profit',
            palette='Set2',
            ax=axes[1])
axes[1].set_title('Operating Profit by Sales Method')
axes[1].set_xlabel('Sales Method')
axes[1].set_ylabel('Operating Profit ($)')

plt.tight_layout()
plt.savefig('chart_sales_method.png', dpi=150)
plt.show()
```

```python
# تجميع المبيعات حسب السنة والشهر
monthly_data = (df.groupby(['Year', 'Month'])['Total Sales']
                .sum()
                .reset_index()
                .sort_values(['Year', 'Month']))

print(monthly_data.head(10))
```

```python
fig, ax = plt.subplots(figsize=(14, 5))

for year, grp in monthly_data.groupby('Year'):
    ax.plot(grp['Month'], grp['Total Sales'] / 1e6,
            marker='o', label=str(year))

ax.set_title('Monthly Sales Trend by Year')
ax.set_xlabel('Month')
ax.set_ylabel('Total Sales (Million $)')
ax.set_xticks(range(1, 13))
ax.set_xticklabels(['Jan','Feb','Mar','Apr','May','Jun',
                    'Jul','Aug','Sep','Oct','Nov','Dec'])
ax.legend(title='Year')

plt.tight_layout()
plt.savefig('chart_monthly_trend.png', dpi=150)
plt.show()
```

```python
# تجميع البيانات حسب الـ Retailer
retailer_data = (df.groupby('Retailer')[['Total Sales', 'Operating Profit']]
                 .sum()
                 .sort_values('Total Sales', ascending=False)
                 .reset_index())

print(retailer_data)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

x = range(len(retailer_data))
w = 0.4

ax.bar([i - w/2 for i in x], retailer_data['Total Sales'] / 1e6,
       width=w, label='Total Sales', color='#4C72B0')

ax.bar([i + w/2 for i in x], retailer_data['Operating Profit'] / 1e6,
       width=w, label='Operating Profit', color='#DD8452')

ax.set_xticks(list(x))
ax.set_xticklabels(retailer_data['Retailer'], rotation=15)
ax.set_ylabel('Sales (Million $)')
ax.set_title('Total Sales & Operating Profit by Retailer')
ax.legend()

plt.tight_layout()
plt.savefig('chart_retailers.png', dpi=150)
plt.show()
```

```python
pivot = df.pivot_table(
    values='Total Sales',
    index='Region',
    columns='Sales Method',
    aggfunc='sum'
) / 1e6

fig, ax = plt.subplots(figsize=(9, 5))

sns.heatmap(pivot,
            annot=True,
            fmt='.1f',
            cmap='YlOrRd',
            linewidths=0.5,
            ax=ax)

ax.set_title('Sales by Region & Sales Method (Million $)')
ax.set_xlabel('Sales Method')
ax.set_ylabel('Region')

plt.tight_layout()
plt.savefig('chart_heatmap.png', dpi=150)
plt.show()
```

```python
fig, ax = plt.subplots()

scatter = ax.scatter(
    df['Price per Unit'],
    df['Operating Profit'] / 1e3,
    c=df['Units Sold'],
    cmap='viridis',
    alpha=0.5,
    s=40
)

plt.colorbar(scatter, ax=ax, label='Units Sold')
ax.set_title('Price per Unit vs Operating Profit')
ax.set_xlabel('Price per Unit ($)')
ax.set_ylabel('Operating Profit (Thousand $)')

plt.tight_layout()
plt.savefig('chart_scatter.png', dpi=150)
plt.show()
```
